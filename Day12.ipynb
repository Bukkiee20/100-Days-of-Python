```python
import random as rd

def easy():
    attempts = 10
    print(f"You have {attempts} attempts remaining to guess the number.")
    random = rd.randint(1,100)
    while attempts > 0:
        guess = int(input("Make a guess:"))
        print(random)
        print(guess)
        if guess > random:
            print("Too high.")
            print("Guess again.")
            attempts -= 1
            print(f"You have {attempts} attempts remaining to guess the number.")

        elif guess < random:
            print("Too low")
            print("Guess again.")
            attempts -= 1
            print(f"You have {attempts} attempts remaining to guess the number.")
        else:
            print(f"You got it! The answer was {random}")
            break
            
    if attempts == 0:
        print("You lose")
        
def hard():
    attempts = 5
    print(f"You have {attempts} attempts remaining to guess the number.")
    random = rd.randint(1,100)
    while attempts > 0:
        guess = int(input("Make a guess:"))
        print(random)
        print(guess)
        if guess > random:
            print("Too high.")
            print("Guess again.")
            attempts -= 1
            print(f"You have {attempts} attempts remaining to guess the number.")

        elif guess < random:
            print("Too low")
            print("Guess again.")
            attempts -= 1
            print(f"You have {attempts} attempts remaining to guess the number.")
        else:
            print(f"You got it! The answer was {random}")
            break
            
    if attempts == 0:
        print("You lose")
        
print("Welcome to the Number Guessing Game!")
print("I am thinking of a number between 1 and 100")
level = input("Choose a difficulty. Type 'easy' or 'hard':")

if level == "easy":
    easy()
else:
    hard()     
```

```python
EASY_LEVEL_ATTEMPTS = 10
HARD_LEVEL_ATTEMPTS = 5


def check_answer(guess, random, attempts):
    if guess > random:
        print("Too high.")
        return attempts - 1
       
    elif guess < random:
        print("Too low")
        return attempts - 1
        
    else:
        print(f"You got it! The answer was {random}")

        
def set_difficulty():
    level = input("Choose a difficulty. Type 'easy' or 'hard':")
    if level == "easy":
        return EASY_LEVEL_ATTEMPTS
    else:
        return HARD_LEVEL_ATTEMPTS

def game():
    print("Welcome to the Number Guessing Game!")
    print("I am thinking of a number between 1 and 100")
    random = rd.randint(1,100)

    attempts = set_difficulty()

    guess = 0
    while guess != random:
        print(f"You have {attempts} attempts remaining to guess the number.")
        guess = int(input("Make a guess:"))
        attempts = check_answer(guess, random, attempts) 
        
        if attempts == 0:
            print("You've run out of guesses, you lose")
            return
        elif guess != random:
            print("Guess again")
        
game()
```
